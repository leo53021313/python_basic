```python
# 建立 function 1 -> 沒有 return值 也沒有參數
# 某個物件才能使用的 function -> method 例如 list的.append() 是 list 的 method

def greeting():
    print("Hello World!")

greeting()
```

```python
# 建立 function 2 -> 有 return值 沒有參數
def get_greeting_str():
    msg = "Hello" + "World!"
    return msg

print(get_greeting_str())
#String 的 method 會 return值 -> 所以要有變數去接他
#List 的 method 沒有 return值 -> 而是修改原本的list
```

```python
# 建立 function 3 -> 有參數
def greeting(name):
    print(f"Hello, {name}")

greeting("Leo")
greeting("Jack")

```

```python
# 建立 function 4 -> 預設參數值

def greeting(name, greeting="Hello"): # 要預設參數值要放在後方
    print(f"{greeting}, {name}")

greeting("Leo") # greeting 有預設參數值， 可以只給一個參數
greeting("Leo", "Good Night")
greeting(greeting="Good Night", name="Leo") # 指定參數

```

```python
# 改寫 Module2 practice

#滿2000 折 200, 滿 4000 折 400, 滿6000 折 600.....
def discount(total):
    if total >= 2000:
        total -= 200*(total//2000)
    return total
    
print(discount(4000))
```

```python
customer_cost = [3000, 5000, 1999, 12000, 9999, 99999]
for cost in customer_cost:
    print(discount(cost))
```

```python
def add(*args, **kwargs):
    print(args)
    print(kwargs)
    total = 0
    for i in args:
        total += i
    return total

total = add(3, 5, 10, 20, 100, unit="元")
print(total)
```
