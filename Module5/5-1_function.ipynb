```python
# е»әз«Ӣ function 3 -> жңүеҸғж•ё
def greeting(name):
    print(f"Hello, {name}")

greeting("Leo")
greeting("Jack")

```

```python
# е»әз«Ӣ function 4 -> й җиЁӯеҸғж•ёеҖј

def greeting(name, greeting="Hello"): # иҰҒй җиЁӯеҸғж•ёеҖјиҰҒж”ҫеңЁеҫҢж–№
    print(f"{greeting}, {name}")

greeting("Leo") # greeting жңүй җиЁӯеҸғж•ёеҖјпјҢ еҸҜд»ҘеҸӘзөҰдёҖеҖӢеҸғж•ё
greeting("Leo", "Good Night")
greeting(greeting="Good Night", name="Leo") # жҢҮе®ҡеҸғж•ё

```

```python
# ж”№еҜ« Module2 practice

#ж»ҝ2000 жҠҳ 200, ж»ҝ 4000 жҠҳ 400, ж»ҝ6000 жҠҳ 600.....
def discount(total):
    if total >= 2000:
        total -= 200*(total//2000)
    return total
    
print(discount(4000))
```

```python
customer_cost = [3000, 5000, 1999, 12000, 9999, 99999]
for cost in customer_cost:
    print(discount(cost))
```
