條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
num = 100
if num > 50:
    print("數字大於 50")
```

##### if...else...

```python
num = 30

if num>50:
    print("數字大於 50")
else:
    print("數字沒有大於 50")
```

##### if...elif... (else)

```python
fruit = "Apple"

if fruit == "Apple":
    print("是蘋果")
elif fruit == "Banana":
    print("是香蕉")
elif fruit == "Grape":
    print("是葡萄")
elif fruit == "Apple":
    print("第二次遇到蘋果")
else:
    print("所有條件不符合")
```

```python
fruit = "Peach"
if fruit in ["Apple", "Banana", "Grape"]:
    print("有在名單內")
else:
    print("沒在名單內")
```
