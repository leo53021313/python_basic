##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''

texts = "contents"
for char in texts:
    print(char)
    if char == "t":
        break
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
texts = "contents"
for char in texts:
    if char == "t":
        continue
    print(char)
```

### 小試身手
計算 1 到 100 之間的和，如果值超過500就停止執行，打印出剛好超過 500 的那個數字

```python
total = 0
for i in range(1, 101):
    total = total + (i + 1)
    print(total + 1)
    if total > 500:
        break
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python
total = 0
for i in range(1, 101):
    if i % 5 == 0 or i % 2 == 0:
        continue
    total += i

print(total)
```

```python
total = 0
for i in range(1, 101, 2):
    if i % 5 == 0:
        continue
    total += i
print(total)
```
