比較運算子 (用 if 判斷)

```python
# ==
# !=
# >
# <
# >=
# <=

print(5==5)
print(5!=5)
print(3<10)
print(9>10)
print(10>=10)
print(9<=10)
```

邏輯運算子 and / or / not

```python
username = "Leo"
password = 123456

if username == "Leo" and password == 123456: #所有條件都要符合
    print("登入成功")


if username == "Leo" or password == 12345: #只需其中一個條件符合
    print("帳號或密碼正確")

if not username == "Allen":
    print("條件為 False 才會執行")
```
