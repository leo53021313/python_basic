百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
total = 2000
event_price = 2000
discount = 200

if total >= event_price:
    print(total - discount)
```

滿2000 折 200, 滿 4000 折 400, 滿6000 折 600.....

```python
# 算術運算子

total = 7000
event_price = 2000
discount = 200
# print(total % event_price)

if total >= event_price:
    print(total - (total // event_price) * discount)
```
