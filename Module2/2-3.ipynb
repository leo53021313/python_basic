整數 (int)、浮點數 (float)

```python
x = 10
y = 5.5
print(x + y)
xyz = 20
print(x + y + xyz)
```

布林值 (boolean)

```python
open = False
print(open)
print(type(open))
```

字串 (string)

```python
name = "Leo"
print(name)
```

列表 (list)

元組 (tuple)

```python
nums = (1, 2, 3)
print(nums)
```

集合 (set) [不能重複]

```python
nums = {1, 2, 3}
print(nums)
print(type(nums))
```

字典 (dict)

```python
user_info = {"name": "Leo", "age": 25}
print(user_info)
print(type(user_info))
```

```python
# 創建一個空的 dict
data = []
print(data)
print(type(data))

# 空的dict而不是set
info = {}
print(info)
print(type(info))
```

```python
# 創建一個空的 set
data = set()
print(data)
print(type(data))
```
