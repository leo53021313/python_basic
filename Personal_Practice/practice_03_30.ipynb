## 終極挑戰二：🚀 列表推導式 (List Comprehension) 特訓
列表推導式的語法是：`[運算式 for 變數 in 範圍 if 條件]`。
請使用**「一行程式碼 (列表推導式)」**完成以下任務：

找出 `1 到 30` 之間（包含 1 和 30）：
1. 能夠被 3 整除 (`%`)
2. 但**不能**被 2 整除（也就是只要奇數）
3. 將符合條件的數字**平方 (`**`)** 後放入串列中。

**Hint:**

## 終極挑戰四：🛡️ 質數搜尋雷達 (迴圈控制 break/continue)
質數是「只能被 1 和自己整除的大於 1 的整數」。
請結合 `while` 迴圈、`for` 迴圈與 `break`，找出**前 5 個質數**，並存入 List。

**任務步驟：**
1. 建立空串列 `primes =[]`
2. 檢查的數字 `num` 從 2 開始。
3. 當 `len(primes) < 5` 時，執行 `while` 迴圈。
4. 在內部寫一個 `for` 迴圈，用 `range(2, num)` 檢查是否有數字能整除 `num`。
5. 如果發生整除，代表不是質數，使用 `break` 結束內迴圈。
6. 如果內迴圈跑完都沒有觸發 `break`，將 `num` 加入 `primes`（提示：可以使用 `for...else...` 語法，或者用一個布林值變數 `is_prime = True` 來記錄）。
7. 別忘了 `num += 1`！

**Hint:**
* 預期輸出：`[2, 3, 5, 7, 11]`

```python
primes =[]
num = 2

# TODO: 完成尋找前 5 個質數的邏輯
while len(primes) < 5:
    is_prime = True
    
    # 檢查是否能被 2 到 num-1 之間的數整除
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
            
    if is_prime:
        primes.append(num)
        
    num += 1

print("前五個質數:", primes)
```

## 終極挑戰五：🐉 大魔王綜合流水線 (String, Dict, Comprehension, List Methods)
在真實開發中，你常會拿到格式混亂的字串，請把所學的全部結合起來清理它！

**任務：**
給你一段包含姓名與成績的混亂字串 `raw_data`：
1. 先將字串用 `,` 切割成一個 List。
2. 建立一個字典 `scores_dict`。使用迴圈將剛才 List 的資料分離：
   - 名字：去除空白、轉為全大寫 (`upper`)。
   - 成績：轉為整數 `int()`。
   *(提示：切割出的元素如 `" ALICE-85 "`，可以再用 `.split('-')` 拆成名字和成績)*
3. **魔王關卡**：使用**列表推導式 (List Comprehension)** 從 `scores_dict` 的中，挑選出「成績 >= 60」的名字存成一個 List。*(提示：利用 `scores_dict` 取值進行判斷)*
4. 將及格名單進行字母排序 (`sort`)。
5. 最後，使用 `join()` 將及格名單用 `" & "` 連接成一個字串。

**Hint:**
* 預期輸入：`"  ALICE-85 ,bOB-42, cHarliE-90 , DAvid-55, eVe-100  "`
* 預期字典：`{'ALICE': 85, 'BOB': 42, 'CHARLIE': 90, 'DAVID': 55, 'EVE': 100}`
* 預期 List Comprehension 結果：`['ALICE', 'CHARLIE', 'EVE']`
* 最終預期輸出：`"ALICE & CHARLIE & EVE"`

```python
raw_data = "  ALICE-85 ,bOB-42, cHarliE-90 , DAvid-55, eVe-100  "

# 1. 根據逗號切割字串 (變成 list)
data_list = raw_data.split(",")

# 2. 建立成績字典
scores_dict = {}
for item in data_list:
    # 先用 '-' 切割，分離名字和分數
    parts = item.split("-")
    
    # 清洗名字 (去空白、轉大寫)
    name = parts[0].strip().upper()
    
    # 將分數轉為整數
    score = int(parts[1])
    
    # 存入字典
    scores_dict[name] = score

print("整理後的字典:", scores_dict)

# 3. 使用 List Comprehension 找出及格 (>= 60) 的名字
# 提示: [name for name in ["ALICE", "BOB", ...] if 條件]
# 你可以從 scores_dict 提取資料來判斷

passed_names = [name for name in scores_dict if scores_dict[name] >= 60]
print("及格名單:", passed_names)

# 4. 排序及格名單
passed_names.sort()

# 5. 用 " & " 拼接為字串
final_string = " & ".join(passed_names)
print("最終結果:", final_string)
```

4. 檢查傳入的 `amount` 是否大於餘額：
   - 如果大於餘額，印出「餘額不足」。
   - 如果足夠，將餘額扣除 `amount`，並印出「提款成功，剩餘：...」。

**Hint:**
* 執行 `withdraw(2000)` 預期輸出：提款成功，剩餘：3000
* 執行 `withdraw(4000)` 預期輸出：餘額不足
* 最終 `account_balance` 應為 3000。

```python
account_balance = 5000

# 定義提款函式
def withdraw(amount):
    global account_balance
    
    # TODO: 判斷餘額並扣款
    if (amount) > account_balance:
        print("餘額不足")
    else:
        account_balance -= amount
        print(f"提款成功，剩餘: {account_balance}")

# 測試區 (請勿更動)
print(f"初始餘額: {account_balance}")
withdraw(2000)
withdraw(4000)
print(f"最終餘額: {account_balance}")
```

3. 實作 `remove_item(item_name)` 函式：利用字典的 `.pop(key)` 方法移除商品。請使用 `try-except` 處理如果移除了不存在的商品時所產生的 `KeyError`，並印出「購物車沒有這個商品」。
4. 實作 `clear_cart()` 函式：在函式內使用 `global` 關鍵字，將 `cart` 重新賦值為空字典 `{}`。

**預期執行結果：**
應印出：`購物車沒有這個商品`
最後的 cart 應為：`{}`

```python
cart = {}

def add_item(item_name, qty):
    global cart
    if cart.get(item_name) is None:
        cart.update({item_name: qty})
    else:
        cart[item_name] += qty

def remove_item(item_name):
    global cart
    try:
        cart.pop(item_name)
    except Exception:
        print("購物車沒有這個商品")

def clear_cart():
    global cart
    cart.clear()


# --- 測試區 (請勿更動) ---
add_item("Apple", 3)
add_item("Banana", 2)
add_item("Apple", 2)  # Apple 應該變成 5
remove_item("Banana")
remove_item("Orange") # 應該觸發 except，印出「購物車沒有這個商品」
print("清空前的購物車:", cart)

clear_cart()
print("清空後的購物車:", cart)
```

這裡有 A 店和 B 店的商品 ID 清單。商品 ID 的格式都是 `"類別-編號"`。
**任務：**
1. 將兩個 List 轉換為集合，找出**兩家店都有賣的商品**（交集）。
2. 使用**一行列表推導式 (List Comprehension)**，從上一步的交集結果中：
   - 使用 `.split("-")` 將字串切開，只取出後面的「編號」。
   - 將編號轉換為整數 (`int`)。
3. 將產生的整數編號 List 進行**降冪排序**（由大到小）。
4. 使用**切片 (Slicing)** 取出最大的前 2 個編號。

**預期輸出結果：**
`[707, 303]`

```python
store_a =["A-101", "B-202", "C-303", "D-404", "E-707"]
store_b =["C-303", "X-999", "A-101", "F-505", "E-707"]

# 1. 找交集
common_items = set(store_a) & set(store_b)

# 2. 列表推導式：取出數字並轉整數
common_numbers = [int(id.split("-")[1]) for id in common_items]

# 3. 排序 (降冪)
common_numbers.sort(reverse=True)

# 4. 切片取前 2 名
top_2_numbers = common_numbers[:2]

print("最大兩個共同編號:", top_2_numbers)
```

## 終極挑戰十四：🏫 結算學生成績 (List of Dicts, Dict Methods, Lambda)
期末考結束了，有些學生因為請假沒有成績（字典裡沒有 `"final"` 這個 Key）。
**任務：**
1. 使用迴圈跑過每一位學生的字典。
2. 使用 `.get()` 方法取得期末考 `"final"` 的成績，如果沒有這個 Key，請預設給 `0` 分。
3. 計算每個人的平均成績：`(midterm + final) / 2`，並將算出的平均分數**新增回該學生的字典裡**，Key 命名為 `"avg"`。
4. 使用 `.sort()` 搭配 `key=lambda ...`，依照剛算出來的 `"avg"` 分數，將學生名單**由高到低（降冪）排序**。

**預期第一名：**
`{'name': 'Charlie', 'midterm': 88, 'final': 92, 'avg': 90.0}`

```python
students =[
    {"name": "Alice", "midterm": 70, "final": 80},
    {"name": "Bob", "midterm": 60}, # 缺考 final
    {"name": "Charlie", "midterm": 88, "final": 92},
    {"name": "David", "midterm": 90, "final": 85}
]

# 1~3. 迴圈處理、取得預設值、新增 avg
for student in students:
    if student.get("final") is None:
        student["final"] = 0

for student in students:
    student["avg"] = (student["midterm"] + student["final"])/2

# 4. 依照 avg 降冪排序 (使用 lambda)
# TODO: students.sort(...)
students.sort(key = lambda x: x["avg"],reverse=True)


print("成績排名:")
for s in students:
    print(s)
```

## 終極挑戰十五：🔥🔥 巢狀資料大搜救 (Nested Loops & Exception Handling)
你的氣象感測器回傳了三天的溫度資料 `sensor_data`。這是一個 **「裝著 List 的 List，裡面又包著字典」** 的恐怖巢狀結構！有些感測器還壞掉回傳了無效的文字。
**任務：**
1. 建立一個空串列 `valid_temps =[]`。
2. 你必須寫一個**雙重迴圈**：外層迴圈跑每一天，內層迴圈跑每天裡面的字典資料。
3. 嘗試把字典裡的 `"temp"` 數值轉為浮點數 `float()`，並存入 `valid_temps`。
4. 遇到無法轉換的雜訊（例如 `"N/A"`, `"Error"`），請用 `try-except` 接住並用 `continue` 忽略它。
5. 最後，請使用串列的 `.sort()` 排列 `valid_temps`，並用切片取出這三天內的**最高溫**與**最低溫**。

**預期輸出結果：**
有效溫度: `[21.5, 22.0, 23.1, 25.5, 26.0]`
最低溫: `21.5`
最高溫: `26.0`

```python
#[ 第一天資料, 第二天資料, 第三天資料 ]
sensor_data = [[{"temp": "25.5"}, {"temp": 26}],
               [{"temp": "N/A"}, {"temp": "23.1"}],
               [{"temp": 22}, {"temp": "Error"}, {"temp": "21.5"}]
]

valid_temps = []
for daily_info in sensor_data:
    for info in daily_info:
        print(info)
        try:
            valid_temps.append(float(info["temp"]))
        except Exception:
            continue

valid_temps.sort()
print(f"有效溫度: {valid_temps} 最低溫: {valid_temps[0]} 最高溫: {valid_temps[-1]} ")
```
