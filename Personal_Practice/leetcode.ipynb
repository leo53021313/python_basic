```python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        if numRows == 0:
            return []
        if numRows == 1:
            return [1]
        
        output_list = []
        inside_list = []
        for i in range(numRows):
            inside_list.clear()
            for j in range(i):
                print(i)
                inside_list.append(i)
            output_list.append(inside_list)

        return output_list
    
s = Solution()
print(s.generate(2))
```
