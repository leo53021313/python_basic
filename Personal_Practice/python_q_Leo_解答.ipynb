### 有一筆學生資料如下:

```python
students_grade = [
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 65, "英文": 40, "數學": 30},
]
```

##### 1. 印出所有學生姓名

```python
print(", ".join([student["name"] for student in students_grade]))
```

##### 2. 印出 `Pochita` 的數學成績

```python
print([student["數學"] for student in students_grade if student["name"] == "Pochita"])
```

#### 3. 計算每位學生的總分
輸出格式  
`Phochita 總分: xxx 分`

```python
for student in students_grade:
    print(f"{student["name"]} 總分: {student["國文"] + student["英文"] + student["數學"]} 分")
```

#### 4. 計算每位學生的平均分數
輸出格式  
`Pochita 平均分數: xxx 分`

```python
for student in students_grade:
    print(f"{student["name"]} 平均分數: {(student["國文"] + student["英文"] + student["數學"])/3:.2f} 分")
```

#### 5. 找出數學最高分的學生

```python
highest_math = 0
highest_student = ""
for student in students_grade:
    if student["數學"] > highest_math:
        highest_math = student["數學"]
        highest_student = student["name"]

print(highest_student)
```

#### 6. 計算全班國文平均

```python
total = 0
for student in students_grade:
    total += student["國文"]

print(total/len(students_grade))
```

#### 7. 找出不及格的學生（任一科 < 60）

```python
bad_student = []
for student in students_grade:
    # print(student)
    for key, value in student.items():
        try:
            if int(value) < 60:
                bad_student.append(student["name"])
                break
        except Exception:
            continue

print(bad_student)
```

#### 8. 新增「總分」到每個 dict
例如 :  
`{"name": "Pochita", "國文": 80, ..., "總分": 245}`

```python
for student in students_grade:
    student["總分"] = student["國文"] + student["英文"] + student["數學"]
    print(student)
```

#### 9. 找出三科都 > 80 的學生

```python
for student in students_grade:
    if student["國文"] > 80 and student["英文"] > 80 and student["數學"] > 80:
        print(student["name"])
```

#### 10. 將資料轉成以下格式 (dict of dict)

```python
"""
{
    "Denji": {"國文": 30, "英文": 10, "數學": 20},
    .....
}
"""
students_grade_dict = {}
for student in students_grade:
    students_grade_dict.update({
        student["name"]: {
            "國文":student["國文"],
            "英文":student["英文"],
            "數學":student["數學"]
        }
    })

print(students_grade_dict)
```
