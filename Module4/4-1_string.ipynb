```python
#String 可以使用index

name = "Leoooooo"
print(name[0])
```

```python
# 字串變數初始化
name ="Alex"
print(name)

name = "Jarvis"  # 重新赴值而不是 不可變
print(name)
```

```python
#字串可以相加
name1 = "Leo"
name2 = "Jack"
name3 = name1 + " " + name2
print(name3)
```

```python
# 替換字串 str.replace(<old>, <new>)
'''

將 string 中的 str1 替換成 str2
'''
# Alex -> Allen
name = "Alex"
new_name = name.replace("ex", "len")
print(new_name)
```

```python
# 去除兩側空格 .strip()
'''

去除字串 string 左、右兩邊的空格
'''
name = "      Hello World       !     "
new_name = name.strip()
print(name)
print(new_name)


data = "XXTeslxaXXXXX"
print(data)
data = data.strip("X")
print(data)

data = "XXTeslaXXXXX"
print(data)
data = data.lstrip("X") # 去除右邊 rstrip()
print(data)
```

```python
# 字串小寫變大寫
'''

將字串 string 裡的字母全部改成小寫 .lower()
'''
print("Jack".lower())


'''

將字串 string 裡的字母全部改成大寫 .upper()
'''

print("Hello World!".upper())
print("Hello 你好嗎".upper())
```

```python
# 字串分割成 list str.split()
greeting = "hello world! hello John"

data = greeting.split("! ")
print(data)

#字串合併 list -> 字串 str.join(<iterable>)
greeting = "__".join(data)
print(greeting)

greeting = "++".join(("你好", "我很好"))
print(greeting)
```
