```python
# f-string (又稱作 formatted string literals, version >= 3.6)
name = "Jarvis"
age = 18
msg = f"My name is {name} and I am {age} years old"
print(msg)
```

```python
# 靠左對齊、靠右對齊

name = "Jarvis"
age = 18
msg = f"My name is [{name:<10}] and I am [{age:>10}] years old"
print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)

name = "Jarvis"
age = 18
msg = f"My name is [{name:<010}] and I am [{age:>010}] years old"
print(msg)
```

```python
# 置中對齊
name = "Jarvis"
age = 18
msg = f"My name is [{name:^10}] and I am [{age:^010}] years old"
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)

name = "Jarvis"
age = 18
msg = f"My name is [{name:0^10.2}] and I am [{age:X>10}] years old"
print(msg)
```

```python
user_id = "A123456789"
result = f"{user_id:X<10.4}"
print(result)
```
