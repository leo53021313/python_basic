##### slice 透過 [ : ] 進行分割
`[n:m]` 走訪 n 到 m - 1 的元素

```python
# 初始化一個字串
data = "Hello,World!"
print(data)
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
data = "Hello,World!"
print(data[0:5])
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
data = "Hello,World!"
print(data[:5]) #從頭不需要寫0
```

```python
# 從 6 取到尾
data = "Hello,World!"
print(data[6:])
```

```python
# 從 7 取到 9（不包含第 9 個元素，實際為 7 ~ 8）
data = "Hello,World!"
print(data[7:9])
```

##### 使用負號

```python
# 從倒數第 10 取到倒數第 7（不包含倒數第 7 的元素）

data = "Hello,World!"
print(data[-10:-7])
```

```python
# 從倒數第 6 取到尾

data = "Hello,World!"
print(data[-5:])
```

```python
# 從頭取到倒數第 7（不包含倒數第 7 的元素）

data = "Hello,World!"
print(data[:-7])
```

```python
# 也可以使用在 list
nums = [i for i in range(1, 21)]
print(nums)

print(nums[:5])

# 搭配 step [n:m:step]

print(nums[:5:2])
print(nums[:5:-1])
```

```python
# 實做練習 將每個單字改成大寫開頭
string04 = "hello world! hello john"

data = string04.split( )
new_data = ""
print(data)
for i in data:
    new_data += i.replace(i[0], i[0].upper()) + " "

print(new_data.strip())

#老師答案:
print(data)
result = ""
for i in data:
    result += i[0].upper() + i[1:] + " "
    # print(i[0].upper() + i[1:])
print(result)


```
