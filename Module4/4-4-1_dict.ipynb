```python
# 初始化 dict
fruit_price = {"apple": 100, "orange": 20, "tomato": 50}
print(fruit_price)
```

```python
#當資料多 建議的寫法:
fruit_price = {
    "apple": 50,
    "orange": 30,
    "tomato": 100
}

print(fruit_price)
```

```python
# 印出 蘋果的價格 -> 取得某個key 的 value
fruit_price = {"apple": 100, "orange": 20, "tomato": 50}
print(fruit_price["apple"])
```

```python
# 修改 橘子的價格
fruit_price["orange"] = 200
print(fruit_price)
```

```python
# 刪除 番茄

fruit_price.pop("tomato")
print(fruit_price)
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name": "Leo",
    "age": 30,
    "info": {
        "nickname": "batman",
        "role": "Pochita",
        "phone_nums" : [
            "0912345678",
            "0987654321"
        ]
    }
}
```

```python
# 取得 nickname
print(user_info["info"])
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"])
print(user_info["info"]["phone_nums"])
print(user_info["info"]["phone_nums"][0]) # 取得第一筆電話號碼
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    "apple": {
        "price": 100,
        "amount": 3,
        "color": [
            "green",
            "red"
        ]
    },
    "orange": {
        "price": 50,
        "amount": 5,
        "color": [
            "orange"
        ]
    },
    "tomato": {
        "price": 80,
        "amount": 10,
        "color": [
            "red"
        ]
    }
}

print(fruit_dict)

# 格式都要一樣

fruit_dict2 = {
    "kiwi": {
        "price": 120,
        "amount": 9,
        "color": [
            "yellow",
            "green"
        ]
    }
}

print(fruit_dict2)
```

##### dict 的合併

```python
# dict 的合併 .update()
fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint

pprint(fruit_dict, sort_dicts=False) #<-- sort_dicts 預設true會打亂dict的順序
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''
pprint(fruit_dict, sort_dicts=False)
pprint(fruit_dict["apple"], sort_dicts=False)
pprint(fruit_dict["apple"]["color"], sort_dicts=False)
fruit_dict["apple"]["color"].append("yellow")
pprint(fruit_dict, sort_dicts=False)
```

```python
pprint(fruit_dict["tomato"]["color"], sort_dicts=False)
fruit_dict["tomato"]["color"].append("green")
pprint(fruit_dict, sort_dicts=False)
```

```python
pprint(fruit_dict["tomato"]["color"], sort_dicts=False)
new_color = "green"
if new_color not in fruit_dict["tomato"]["color"]:
    fruit_dict["tomato"]["color"].append("green")
pprint(fruit_dict, sort_dicts=False)
```
