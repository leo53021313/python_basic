```python
# 初始化一個 list
# list [<value>, <value>, ...]
list1 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
print(list1)
```

```python
# 新增元素在 list 尾端
# list.append(<value>)
list1 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
list1.append(1)
print(list1)
```

```python
# 修改索引位置的元素
# list[<index>]

list1 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
list1[0] = 1
print(list1)
```

```python
# 刪除索引位置的元素

list1 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
list1.pop(1)
print(list1)
```

```python
# 新增元素 9 在指定索引 1，其餘元素往後移
list1 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
list1.insert(1, 0)
print(list1)
```

```python
# 移除指定的值
list1 = ["星期一", "星期二", "星期三", "星期四", "星期五", "星期六", "星期日"]
list1.remove("星期二")
print(list1)
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list

nums = []
for i in range(1, 11):
    nums.append(i**2)

print(nums)
```

```python
# enumerate() 同時取得 index 與 value

nums = []
for i in range(1, 11):
    nums.append(i**2)

print(nums)

for index, value in enumerate(nums):
    print(f"index: {index}, 值: {value}")
    if value == 49:
        print(index)
```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
nums = []
for i in range(1, 21):
    nums.append(i)
print(nums)


# 簡寫:
new_list = [i for i in range(1, 21)]
print(nums)
```

```python
nums = []
for i in range(1, 21):
    if i % 5 == 0:
        nums.append(i)
print(nums)

list1 = [i for i in range(1, 21) if i % 5 == 0]
print(list1)
```

```python
grade = 40
if grade < 60:
    grade += 10
else:
    grade += 5

print(grade)
```

```python
# 三元運算
grade = 40
grade = grade + 10 if grade < 60 else grade + 5

print(grade)
```

```python
# 三元運算
grade = 60
print("不及格" if grade < 60 else "及格")
```

```python
grades = [60, 50, 100, 90, 80, 70, 40, 10, 20, 30, 35]

passing = ["及格" if grade >= 60 else "不及格" for grade in grades]

print(passing)

print(passing.count("及格"))
print(passing.count("不及格"))

# passing2 = [grade for grade in grades if grade >= 60 "及格" else "不及格"]
# print(passing2)
```
